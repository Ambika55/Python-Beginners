# Control Flow
Decision Making in programming is similar to decision making in real life. A programming language uses control statements to control the flow of execution of the program based on certain conditions. These are used to cause the flow of execution to advance and branch based on changes to the state of a program.



- if, else, elif

```python
a = 10
b = 10
if a<b:
    print("a is greatest")
```

```python
if a>b:
    print("a is greatest")
else:
    print("b is greatest")
```

```python
if a>b:
    print("a is greatest")
elif b>a:
    print("b is greatest")
else: 
    print("a and b are equal")
```

```python
if a>b:
    if a>0:
        print("positive")
    else:
        print("negative")
```

## Loops 

-  for, while, for-else

```python
i = 1

while i<10:
    print(i)
    i+=1
```

```python
l=["apple","banana","cherry"]

for x in l:
    print(x)

for i in range(1,11,3):
    print(i)

for i in range(10,1,-2):
    print(i)
```

## Break
With the break statement we can stop the loop even if the while condition is true:

```python
for i in range(1,10):
    print(i)
    if i == 5:
        break
```

## Continue
With the continue statement we can stop the current iteration, and continue with the next:

```python
for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

## Pass
Skip the code. Used for no content


```python
if a==10:
    pass

for i in range(1,10):
    pass
```

## Nested Loops
- A nested loop is a loop inside a loop.

- The "inner loop" will be executed one time for each iteration of the "outer loop":



```python
for i in range(1,5):
    for j in range(1,5):
        print(i*j)
```

```python
for i in range(1,3):
    print(i)
else:
    print("end of for")
```

# Functions

Functions are generally the block of codes or statements in a program that gives the user the ability to reuse the same code which ultimately saves the excessive use of memory, acts as a time saver and more importantly, provides better readability of the code.

- A function is a block of code which only runs when it is called.

- You can pass data, known as parameters, into a function.

- A function can return data as a result.



```python
def myfunc():
    print("My Function")

myfunc()
```

```python
def myfunc(x):
    return 2*x

y = myfunc(10)
print(y)
```

```python
def myfunc(*x):
    z=1
    for i in x:
        z*=i
    print(z)
        
    
myfunc(10,20,30)
```

```python
def key(lname,age):
    print("My name is",lname)
    print("My age is",age)
    
key(age=20,lname="Ambika")
# key(20,"Ambika")
```

```python
def key(**k):
    for i,j in k.items():
        print(i,j)

key(age=20,lname="Ambika",location="Delhi")
```

## Lambda Function
- A lambda function is a small anonymous function.

- A lambda function can take any number of arguments, but can only have one expression.

Syntax  
lambda arguments : expression

```python
x = lambda a:a+10
print(x(2))
```

```python
x = lambda a,b,c: a*b*c
print(x(1,2,3))
```

```python
def myfunc(n):
    return lambda a:a*n

ans=myfunc(2)
```

```python
print(ans(10))
```
